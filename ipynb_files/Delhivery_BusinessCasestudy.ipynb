#5. Missing values Treatment & Outlier treatment

```python
# Handle outliers for all numeric columns
def remove_outliers(df, columns):
    df_clean = df.copy()
    for col in columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    return df_clean

df_clean = remove_outliers(df, numeric_columns)
```

```python
# Calculate skewness for numeric columns in df_clean
skewness_clean = df_clean[numeric_columns].skew().sort_values(ascending=False)

print("Skewness of numeric features in df_clean:")
print(skewness_clean)
```

```python
# Hypothesis testing between tot_segment_actual_time and tot_osrm_actual_time

from scipy.stats import ttest_ind
# Perform the t-test
t_stat, p_value = ttest_ind(df3['tot_segment_actual_time'], df3['tot_segment_osrm_time'])

# Print the results
print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis. There is a significant difference between the two variables.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the two variables.")
```

```python

# Filter out rows where Source Name equals Destination Name
df_filtered = df3[df3["s_place"] != df3["d_place"]]

# Group by (Source, Destination) and aggregate metrics
busiest_routes = (
    df_filtered.groupby(["s_place", "d_place"])
    .agg(
        Trip_Count=("trip_uuid", "count"),
        Avg_Distance=("segment_osrm_distance", "mean"),
        Avg_Time_Taken=("segment_actual_time", "mean")
    )
    .sort_values("Trip_Count", ascending=False)
    .head(10)
    .reset_index()
)

busiest_routes = busiest_routes.rename(columns={"s_place": "Source Name", "d_place": "Destination Name"})

print("Busiest Routes:")
print(busiest_routes)

# Plot the busiest routes
plt.figure(figsize=(12, 6))
bars = plt.bar(
    busiest_routes.apply(lambda x: f"{x['Source Name']} → {x['Destination Name']}", axis=1),
    busiest_routes["Trip_Count"],
    color="skyblue"
)

# Add annotations for avg distance and time taken
for bar, distance, time in zip(bars, busiest_routes["Avg_Distance"], busiest_routes["Avg_Time_Taken"]):
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # Center the text horizontally
        bar.get_height() + 0.1,  # Place the text slightly above the bar
        f"{distance:.1f} km\n{time:.1f} min",  # Text to display
        ha="center",  # Horizontal alignment
        fontsize=10,
        color="black"
    )

# Add title and labels
plt.title("Busiest Routes by Trip Count with Avg Distance and Time", fontsize=14)
plt.xlabel("Source → Destination", fontsize=12)
plt.ylabel("Trip Count", fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```
