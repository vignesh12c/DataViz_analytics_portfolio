```python
# Generate the QQ plot
from scipy.stats import probplot

data = df['count']

fig, ax = plt.subplots(figsize=(8, 6))

probplot(data, dist="norm", plot=ax)
ax.set_title('QQ Plot df')
ax.set_xlabel("Theoretical Quantiles")
ax.set_ylabel("Sample Quantiles")

# Show plot
plt.show()
```

```python
# Perform Shapiro-Wilk test
from scipy.stats import shapiro

stat, p = shapiro(data)

# Display results
print(f"Shapiro-Wilk Test Statistic: {stat}")
print(f"P-Value: {p}")

# Interpretation
alpha = 0.05
if p > alpha:
    print("The data appears to be normally distributed (fail to reject H0).")
else:
    print("The data does not appear to be normally distributed (reject H0).")
```

```python
from scipy.stats import levene

s1 = df[df['season'] == 1]['count']
s2 = df[df['season'] == 2]['count']
s3 = df[df['season'] == 3]['count']
s4 = df[df['season'] == 4]['count']

stat, p = levene(s1, s2, s3, s4, center='median')

print(f"Levene's Test Statistic: {stat}")
print(f"P-Value: {p}")

if p > 0.05:
    print("The variances are equal (fail to reject H0).")
else:
    print("The variances are not equal (reject H0).")
```

```python
# Perform one-way ANOVA

# Null: Seaon do not have an effect on cycle count (s1 = s2 = s3 = s4)
#alternate: atleast one season do have an effect (s1 != s2 != s3 != s4)
# significance level = 5%

from scipy.stats import f_oneway

s1 = df[df['season'] == 1]['count']
s2 = df[df['season'] == 2]['count']
s3 = df[df['season'] == 3]['count']
s4 = df[df['season'] == 4]['count']

f_stat, p_value = f_oneway(s1, s2, s3, s4)

print(f"F-statistic: {f_stat:.2f}")
print(f"P-value: {p_value}")
```

P value being close to zero. So season has a strong correlation with count.


```python
# Perform t-test to find if the season having greater count than the other

# Null: Seaon 1 and 2 have same demand for cycle (s2 = s1)
#alternate: season 2 has more demand than that of season 1 (s2 > s1)
# significance level = 5%

from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(s2, s1, alternative= 'greater')

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value}")
```

season 2 has more demand than that of season 1

```python
# Null: Seaon 2 and 2 have same demand for cycle (s3 = s2)
#alternate: season 3 has more demand than that of season 2 (s3 > s2)
# significance level = 5%

from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(s3, s2, alternative= 'greater')

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value}")

if p_value < 0.05:
 print("reject null hypothesis")
else:
 print("fail to reject null hypothesis")
```

 season 3 has more demand than that of season 2


```python
# Null: Seaon 3 and 4 have same demand for cycle (s3 = s4)
#alternate: season 3 has more demand than that of season 4 (s3 > s4)
# significance level = 5%

from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(s3, s4, alternative= 'greater')

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value}")

if p_value < 0.05:
 print("reject null hypothesis")
else:
 print("fail to reject null hypothesis")
```

```python
# Specify the target column for correlation
target_column = 'count'

# Filter float columns
float_columns = df.select_dtypes(include=['float']).columns

# Calculate correlations
pearson_corr = df[float_columns].corrwith(df[target_column], method='pearson')
spearman_corr = df[float_columns].corrwith(df[target_column], method='spearman')

# Combine results
correlation_results = pd.DataFrame({
    'Pearson': pearson_corr,
    'Spearman': spearman_corr
})
correlation_results
```
