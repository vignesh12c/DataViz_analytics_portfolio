# 4. Representing the marginal probability like - what percent of customers have purchased KP281, KP481, or KP781 in a table (can use pandas.crosstab here)

```python
cross_tab = pd.crosstab(df['Product'], df['Gender'], margins=True)
cross_tab
```

```python
marginal_probabilities = cross_tab['All'][:-1] / cross_tab['All']['All']
marginal_probabilities
```

```python
conditional_prob = cross_tab.div(cross_tab.loc['All'], axis=1)
conditional_prob
```

```python
cond_prob = conditional_prob.iloc[:-1, :-1]
cond_prob
```

```python
# Maximum and minimum for Female
max_female_value = cond_prob['Female'].max()  # Max value for Female
max_female_index = cond_prob['Female'].idxmax()  # Coffee type with max Female value

min_female_value = cond_prob['Female'].min()  # Min value for Female
min_female_index = cond_prob['Female'].idxmin()  # Coffee type with min Female value

# Maximum and minimum for Male
max_male_value = cond_prob['Male'].max()  # Max value for Male
max_male_index = cond_prob['Male'].idxmax()  # Coffee type with max Male value

min_male_value = cond_prob['Male'].min()  # Min value for Male
min_male_index = cond_prob['Male'].idxmin()  # Coffee type with min Male value

print(f'Female: Max = {max_female_value} (for {max_female_index}), Min = {min_female_value} (for {min_female_index})')
print(f'Male: Max = {max_male_value} (for {max_male_index}), Min = {min_male_value} (for {min_male_index})')
```

**Conditional Probability (Gender):**



*   Most preferred machine for female is KP281 and least one is KP781
*   Most preferred machine for male is KPKP281 and least one is KP481



```python
cross_tab = pd.crosstab(df['Product'], df['MaritalStatus'], margins=True)
conditional_prob = cross_tab.div(cross_tab.loc['All'], axis=1)
cond_prob = conditional_prob.iloc[:-1, :-1]

# Maximum and minimum for Single
max_single_value = cond_prob['Single'].max()  # Max value for Single
max_Single_index = cond_prob['Single'].idxmax()  # Coffee type with max Single value

min_Single_value = cond_prob['Single'].min()  # Min value for Single
min_Single_index = cond_prob['Single'].idxmin()  # Coffee type with min Single value

# Maximum and minimum for Partnered
max_Partnered_value = cond_prob['Partnered'].max()  # Max value for Partnered
max_Partnered_index = cond_prob['Partnered'].idxmax()  # Coffee type with max Partnered value

min_Partnered_value = cond_prob['Partnered'].min()  # Min value for Partnered
min_Partnered_index = cond_prob['Partnered'].idxmin()  # Coffee type with min Partnered value

print(f'Single: Max = {max_single_value} (for {max_Single_index}), Min = {min_Single_value} (for {min_Single_index})')
print(f'Partnered: Max = {max_Partnered_value} (for {max_Partnered_index}), Min = {min_Partnered_value} (for {min_Partnered_index})')
```

**Conditional Probability (MaritalStatus):**



*   Most preferred machine for single is KP281 and least one is KP781

# Customer Profiling - Categorization of users.

```python
df2 = df.copy()
from sklearn.preprocessing import StandardScaler
numeric_cols = ['Age', 'Income', 'Miles', 'Usage', 'Education']
scaler = StandardScaler()
df2[numeric_cols] = scaler.fit_transform(df2[numeric_cols])
```

```python
df2 = pd.get_dummies(df2, columns=['Product', 'Gender', 'MaritalStatus'], drop_first=True)
```

```python
from sklearn.cluster import KMeans

# Define the number of clusters
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model
df2['cluster'] = kmeans.fit_predict(df2)
```

```python
df2.groupby('cluster').mean()
```

```python

sns.scatterplot(data=df2, x=df2['Age'], y=df2['Income'], s= df2['Usage']*50, cmap='viridis', alpha=0.7, c=df2['cluster'])

plt.title('Age vs Income with Usage as Size (Clustered)')
plt.xlabel('Age')
plt.ylabel('Income')

plt.colorbar(scatter, label='Cluster')

plt.show()
```

**Customer profiling:**


*   cluster 0: low income, young
*   cluster 1: low income, old
*   cluster 2: high income, old



```python
sns.pairplot(df2, hue='cluster', diag_kind='kde', palette='viridis')
plt.show()
```
