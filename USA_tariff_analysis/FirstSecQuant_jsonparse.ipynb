```python
import os
import json
import pandas as pd
from collections import defaultdict

# Configuration
input_folder = "/content/drive/MyDrive/DatawebUSITC/secquantity_2020_2025"  # folder with JSON files
output_folder = "/content/drive/MyDrive/DatawebUSITC/csv_output"
os.makedirs(output_folder, exist_ok=True)

# Storage for parsed data
year_data = defaultdict(list)

# Column headers (19 columns)
columns = [
    "Country", "Year", "District", "Program", "RPC Description", "HTS Number", "Quantity Description",
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

# Loop through each file
for filename in os.listdir(input_folder):
    if filename.endswith(".json") and filename.startswith("secquantity_commodity_"):
        file_path = os.path.join(input_folder, filename)
        try:
            with open(file_path, "r") as f:
                data = json.load(f)

            # Safe navigation through nested JSON
            tables = data.get("dto", {}).get("tables")
            if not tables:
                print(f"Skipping (no tables): {filename}")
                continue

            rows_group = tables[0].get("row_groups")
            if not rows_group or not rows_group[0].get("rowsNew"):
                print(f"Skipping (no rowsNew): {filename}")
                continue

            rows = rows_group[0]["rowsNew"]

            for row in rows:
                values = [entry.get("value", "") for entry in row["rowEntries"]]
                if len(values) == len(columns):
                    year = values[1]
                    year_data[year].append(values)

        except Exception as e:
            print(f"Error processing {filename}: {e}")

# Write combined CSV per year
for year, rows in year_data.items():
    df = pd.DataFrame(rows, columns=columns)
    output_path = os.path.join(output_folder, f"import_secunit_{year}.csv")
    df.to_csv(output_path, index=False)
    print(f"Saved: {output_path}")
```
